9. Write a Python function to calculate Z-scores from a dataset and plot them.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def calculate_z_score(data):
    mean = np.mean(data)
    std_dev = np.std(data)
    z_scores = (data - mean) / std_dev
    return z_scores

def plot_z_scores(z_scores):
    plt.figure(figsize=(8, 5))
    sns.histplot(z_scores, bins=30, kde=True, color='blue', edgecolor='black')
    plt.xlabel('Z-Score')
    plt.ylabel('Frequency')
    plt.title('Z-Score Distribution')
    plt.legend()
    plt.grid(axis="y", linestyle= "--", alpha= 0.7)
    plt.show()


np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=1000)
z_scores = calculate_z_score(data)
plot_z_scores(z_scores)
```

19. Implement hypothesis testing using Z-statistics for a sample dataset.

```python
import numpy as np
from scipy. stats import norm

def z_test(sample_data, population_mean, population_std, alpha=0.05):
    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)
    z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
    z_critical = norm.ppf(1- alpha/2)

    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Z-statistic: {z_stat:.2f}")
    print(f"Critical Z-value: {z_critical:.2f}")

    if abs(z_stat) > z_critical:
        print("Reject the null hypothesis")
    else:
        print("Fail to reject the null hypothesis")

sample_data = np.array([85, 88, 92, 91, 87, 90, 85, 86, 89, 91])

population_mean = 90
population_std = 5

z_test(sample_data, population_mean, population_std)

```

```python
import numpy as np
from scipy.stats import skew, kurtosis

def calculate_skewness_kurtosis(data):
    skewness = skew(data)
    kurt = kurtosis(data)
    return skewness, kurt

np.random.seed(42)
data = np.random.normal(loc=0, scale=1, size=1000)

skewness, kurt = calculate_skewness_kurtosis(data)
print(f"Skewness: {skewness:.4f}")
print(f"Kurtosis: {kurt:.4f}")
```
